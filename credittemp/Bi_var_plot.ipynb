```python
#

def hist_plot(df,vari_col,flag_col,flag_value=[1,0],label=['True','False']):
    import seaborn as sns
    import matplotlib.pyplot as plt
    import numpy as np
    
    sns.set_style("whitegrid")
    
    # 变量的箱须图
    plt.boxplot(df[vari_col])
    plt.title("vari "+vari_col+" box distribute")
    # 安类目分布图
    fig,ax_hist = plt.subplots()
    ax_hist2 = ax_hist.twinx()
    
    # 不同区间的直方图
    ax_hist.set_ylabel("Numbers Count",color="b")
    n,bins,patches = ax_hist.hist(
        [df[df[flag_col] == flag_value[0]][vari_col], 
         df[df[flag_col] == flag_value[1]][vari_col]],
        rwidth = 0.75,
        label = label,
        color = ["g", "b"],
        alpha = 0.75,
        bins=10)
    
#     # 显示绿色部分的数值
#     for a1,b1 in zip(n[0], bins): 
#         ax_hist.text(b1, a1, str(int(a1)))
    
#     # 显示蓝色部分的数值    
#     for a2,b2 in zip(n[1], bins): 
#         ax_hist.text(b2, a2, str(int(a2)))
    
    # 显示不同区间的 GB 占比：
    y= []
    x= []
    i=0  
    for i in range(bins.size-1):
        if ((n[0][i]+n[1][i]))>0:
            y.append(n[0][i]/(n[0][i]+n[1][i]))
        else: y.append(0.5)
        x.append((bins[i]+bins[i+1])/2)
  
    
    ax_hist2.plot(x,y,'rs',x,y,'k')
    ax_hist2.set_ylabel('Green VS All', color='r')
    
    # 显示比例曲线的值
    for a,b in zip(x, y): 
        ax_hist2.text(a, b, str(round(b,3)))
    # 设置其他参数
    plt.xticks(bins) # 坐标轴
    ax_hist.legend()
    fig.set_size_inches(10, 5)
    plt.title("vari "+vari_col+" distribute with classes")
    plt.show()
```

```python
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
y = data["target"][:100]
x = data["data"][:100, 0:4]

samples = pd.DataFrame(
            {"A": x[:, 0].ravel(),
             "B": x[:, 1].ravel(),
             "C": x[:, 2],
             "D": x[:, 3],"label":y[:]})
samples.head()
```

```python
hist_plot(samples,"A","label")
```
