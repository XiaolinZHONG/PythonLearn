```python

import pandas as pd
import numpy as np

def make_jd_samples(df):
	df_type=pd.get_dummies(df.ix[:,"type"])
	# 1 浏览 2 添加购物车 3 从购物车删除 4 下单 5 关注 6 点击
	# 修改数据
	df_type.columns=["view","add","delet","label","fellow","click"]
# 	print(df_type.head())
	df_new=pd.concat([df[["user_id","sku_id","time","model_id","cate","brand"]],df_type],axis=1)
# 	print(df_new.head())
	df_left=df_new.groupby(["user_id","sku_id"])["view","add","delet","fellow","click"].sum().reset_index()
# 	print(df_left.head())
	df_right=df_new[["user_id","sku_id","label"]].groupby(["user_id","sku_id"]).sum().reset_index()
# 	print(df_right.head())
	#检查数据数目
# 	print(df_left.shape,df_right.shape)
	df_2=pd.merge(df_left,df_right,how="inner",on=["user_id","sku_id"])
	print("sample shape:",df_2.shape)
	print(df_2.head())
	return df_2
```

```python
df_buy_2=make_jd_samples(df_buy)
```

```python
df_not_buy_2=make_jd_samples(df_not_buy)
```
